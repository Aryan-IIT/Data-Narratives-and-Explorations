# Question 2

Let us try to define a metric for the "best performing player per season" and "worst perfroming player per season". In order to find this, we can try to find the player with best and worst win to loss ratio for a season.

**Question**: Go through each season and try to find the player with the best and worst win to loss ratio for the given season.

```python
def top_player_generator(season_df, season_number):
  winner_names = season_df["Winner"].unique()
  loser_names = season_df["Loser"].unique()
  player_names_list = set(winner_names).union(loser_names)
  player_names = list(player_names_list)

  player_dict = {name: [0, 0] for name in player_names}
  #0th index denotes the wins, 1st index denotes the losses.

  #9th Column is the name of the winner, 10th column is the name of loser

  row_length = season_df.shape[0]

  for row_number in range(0,row_length):
    row = season_df.iloc[row_number,:]

    loser = row["Loser"]
    winner = row["Winner"]

    player_dict[winner][0]+=1
    player_dict[loser][1]+=1

  win_ratio_dict = {name: player_dict[name][0] / player_dict[name][1] if player_dict[name][1] != 0 else 0 for name in player_names}

  most_effective_player = max(win_ratio_dict, key=win_ratio_dict.get)
  least_effective_player = min(win_ratio_dict, key=win_ratio_dict.get)

  return [ season_number, most_effective_player, player_dict[most_effective_player][0], player_dict[most_effective_player][1],win_ratio_dict[most_effective_player],
          least_effective_player, player_dict[least_effective_player][0],player_dict[least_effective_player][1], win_ratio_dict[least_effective_player] ]
```

```python
column_names_Q2 = ["Season", "Best Performing Player", "Wins", "Losses","Win to Loss Ratio", "Worst Performing Player", "Wins", "Losses", "Win to Loss Ratio"]
df_list_Q2=[]
current = 2000
for season in data_season_frames_season_list:
  df_list_Q2.append(top_player_generator(season, current))
  current+=1
```

```python
df_performace = pd.DataFrame(df_list_Q2, columns = column_names_Q2 )
df_performace
```

Based on the locations provided in the dataset, ATP since 2000 has occured in 120 locations (Cities). Let us try to find the maximum scorer on each location.

**Question:** Find the player that has maximum points on a given location. Consequently, let us try to find the player with a good performance on more than 1 city.

(2005 onwards all datasets have Winner and Loser Points)

(Some cities do not have WPts and LPts so we are dropping them)

```python
atp_locations = df_atp_master["Location"].unique()

df_merged_Q3 = pd.concat([df_atp_2005,
                    df_atp_2006, df_atp_2007, df_atp_2008, df_atp_2009, df_atp_2010, df_atp_2011,
                    df_atp_2012, df_atp_2013, df_atp_2014, df_atp_2015, df_atp_2016, df_atp_2017,
                    df_atp_2018, df_atp_2019, df_atp_2020, df_atp_2021, df_atp_2022, df_atp_2023,
                    df_atp_2024], ignore_index=True)
```

```python
def max_win_points(loc, df_city):
    players_win = df_city["Winner"].unique()
    players_loss = df_city["Loser"].unique()

    union_set = set(players_win).union( set(players_loss) )
    players = list(union_set)

    player_dict = {player: 0 for player in players}

    for player in players:
        winner_df = df_city[df_city["Winner"] == player]  # getting winning points
        player_dict[player] += np.nanmean(winner_df["WPts"])
        #getting mean because points given in WPts is cummulative

        loser_df = df_city[df_city["Loser"] == player]  # getting losing points
        player_dict[player] += np.nanmean(loser_df["LPts"])
        #getting mean because points given in LPts is cummulative

    if len(player_dict)==0:
      return
    else:
      max_key = max(player_dict, key=player_dict.get)

    return [loc, max_key, player_dict[max_key]]
```

```python
best_turf_Q3 = []
column_q3 = ["Location", "Best Player","Mean Points"]
for location in atp_locations:
  df_city = df_merged_Q3[df_merged_Q3["Location"]==location]
  best_turf_Q3.append(max_win_points (location,df_city))

best_turf_Q3 = list(filter(lambda x: x is not None, best_turf_Q3))
```

```python
df_q3 = pd.DataFrame (best_turf_Q3, columns = column_q3)
df_q3.dropna(subset=["Mean Points"])
```

# Question 4

Top 10 earliest debutants of ATP, based on birth year and pro_year in the players data set.

**Question** : Find the top 10 youngest Debutants of ATP.

*Note : As per ATP norms, there is no pre-set age limit to debut in the competition.*


```python
def debut_age_calc(df):
  cols = df.shape[0]
  player_dict = {}

  for i in range(0,cols):
    if df.iloc[i, 2] == 0 or pd.isna(df.iloc[i, 2]) or pd.isna(df.iloc[i, 1]) or df.iloc[i, 1] == 0:
      continue  # Exit the loop if any condition is met
    birthyear = df.iloc[i,1]
    turnedpro = df.iloc[i,2]

    player_dict[df.iloc[i,0]] = turnedpro - birthyear

  sorted_players = sorted(player_dict.items(), key=lambda x: x[1])

  top_10_keys = [item[0] for item in sorted_players[:10]]

  return_list = []

  for i in range(0,len(top_10_keys)):
    return_list.append([top_10_keys[i], player_dict[top_10_keys[i]]])

  return return_list
```

```python
df_filtered_players = df_players[["full_name","birth_year","turned_pro"]]

answer_Q4 = debut_age_calc(df_filtered_players)

pd.DataFrame(answer_Q4,columns=["Player Name","Debut Age (+-1)"])
```

# Question 5

**Assumption** : The columns avgw and avgl are :

Say, player A and B compete, A wins and B loses. A has rank a and B has rank b.

- avgw: This could represent the average (mean) of the win odds offered by multiple bookmakers for the player that won. Let this be avgw_one.

Say, you put 10USD on player A to win, and player A ends up winning. Then, you will gain 10 * avgw_one USD.

- avgl: Similarly, this could represent the average (mean) of the win odds offered by multiple bookmakers for the player that lost. Let this be avgl_one.

Say, you put 10USD on player B to win, and player A ends up winning. Then, you *would have made* 10 * avgl_one USD. But since, your bet failed, you will earn nothing.


Here, wpts and lpts indicate the winning points (points gained by the winner) and lpts indicate the losing points (points gained by the loser).

**Question** : Try to find the co-relation between ranks of the winner and loser, and the odds that are set on them. (Perform this only for the year 2023 for simplicity)

```python
df_atp_2023['Difference_Rank'] = df_atp_2023['LRank'] - df_atp_2023['WRank']
df_atp_2023['Difference_Avg'] = df_atp_2023['AvgL'] - df_atp_2023['AvgW']

corr = df_atp_2023['Difference_Rank'].corr(df_atp_2023['Difference_Avg'])

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Difference_Avg', y='Difference_Rank', data=df_atp_2023, color='green')
plt.title(f'Correlation: {corr:.2f} ({corr * 100:.2f}%)', color='green')
plt.xlabel('Difference in AvgL (Loser Odds) - AvgW (Winner Odds)')
plt.ylabel('Difference in Rankings')
plt.show()

# Inference:
'''
If a higher ranked player (Numerically high relative to opponent) player wins, then his probability
to win the match will be low, but his odds (avgw) will be high for this very reason, because he is not expected to win.
Hence, both avgl-avgw and lrank - wrank will be negative. And vise versa.
Hence, the extent of this co relation is 27%
'''
```

# Question 9

BMI is said to be the body mass index, an essential parameter for fitness. The formula for BMI is weight in kilograms divided by height in meters squared. Let us try to calculate the mean of this for each country.

**Questiont** : Calculating the average BMI of every country's players.

(Additionally, if data(names) was more uniform across players dataset and match dataset, we can map the number of wins of a country and then try to view them against the mean BMI.)

```python
def calculate_bmi(weight_kg, height_cm):
    if weight_kg == 0 or height_cm == 0:
        return
    else:
        height_m = height_cm / 100
        bmi = weight_kg / (height_m ** 2)
        return bmi

df_players['BMI'] = df_players.apply(lambda row: calculate_bmi(row['weight_kg'], row['height_cm']), axis=1)

avg_bmi_by_country = df_players.groupby('flag_code')['BMI'].mean().reset_index()

avg_bmi_by_country_filtered = avg_bmi_by_country[(avg_bmi_by_country['BMI'] != 0) & (~avg_bmi_by_country['BMI'].isnull())]

avg_bmi_by_country_sorted = avg_bmi_by_country_filtered.sort_values(by='BMI')

last_row_index = avg_bmi_by_country_sorted.index[-1] #last country COL is an outlier because a data row has height of player as 3 cm, hence we are dropping this off.

df_final = avg_bmi_by_country_sorted.drop(last_row_index)
df_final.head(10)
```

**Question** : Using heat maps to visualize: player's win perfomace across years, player's season end rank across years and a heatmap of venues played across the years. Let the player be **Andy Murray**

(For simplicity, let us perform this on years 2013, 2014, 2015, 2016, 2017, 2018, 2019)

```python
df_Q10 = pd.concat([df_atp_2013,df_atp_2014,df_atp_2015,df_atp_2016,df_atp_2017,df_atp_2018,df_atp_2019],ignore_index=True)
df_murray = df_Q10[(df_Q10["Winner"]=="Murray A.") | (df_Q10["Loser"]=="Murray A.")]
```

```python
#1. Murray A.'s Win Percentage Over the Years
df_murray['Year'] = pd.to_datetime(df_murray['Date']).dt.year

matches_played = df_murray.groupby('Year').size()
matches_won = df_murray[df_murray['Winner'] == 'Murray A.'].groupby('Year').size()

win_percentage = (matches_won / matches_played).fillna(0)  # Fill NaN with 0 (in case no matches won)

plt.figure(figsize=(8, 4))
sns.heatmap(win_percentage.to_frame(), cmap='Greens', annot=True, fmt=".2%", cbar=False)
plt.title("Murray A.'s Win Percentage Over the Years")
plt.xlabel('Year')
plt.ylabel('Win Percentage')
plt.show()
```

```python
#2. Murray A.'s Ranking Over the Years
df_ranking = df_murray[['Year', 'WRank']]

df_ranking['WRank'] = pd.to_numeric(df_ranking['WRank'], errors='coerce')

df_ranking.dropna(subset=['WRank'], inplace=True)

plt.figure(figsize=(8, 4))
sns.heatmap(df_ranking.pivot_table(index='Year', values='WRank', aggfunc='mean').astype(int),
            cmap='Purples', annot=True, fmt="d", cbar=False)
plt.title("Murray A.'s Ranking Over the Years")
plt.xlabel('Year')
plt.ylabel('Average ATP Ranking')
plt.show()
```

```python
#3. Murray A.'s Performance by Location
plt.figure(figsize=(10, 6))
sns.heatmap(pd.crosstab(df_murray['Year'], df_murray['Location']),
            cmap='Blues', annot=True, cbar=False)
plt.title("Murray A.'s Performance (Matches) by Location")
plt.xlabel('Location')
plt.ylabel('Year')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
